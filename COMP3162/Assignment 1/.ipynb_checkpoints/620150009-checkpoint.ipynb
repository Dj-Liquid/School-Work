```python
from scipy.stats import ttest_ind


houses_1h_ocean = df[df['ocean_prox_int'] == 0]['median_house_value'] #1H < OCEAN
houses_inland = df[df['ocean_prox_int'] == 1]['median_house_value'] #INLAND
houses_island = df[df['ocean_prox_int'] == 2]['median_house_value'] #ISLAND
houses_near_bay = df[df['ocean_prox_int'] == 3]['median_house_value'] #NEAR BAY
houses_near_ocean = df[df['ocean_prox_int'] == 4]['median_house_value'] #NEAR OCEAN

locations = [[houses_1h_ocean,houses_inland,houses_island,houses_near_bay,houses_near_ocean],['1H from Ocean','Inland','Island','Near Bay','Near Ocean']]
t_values = []

for i in range(5):
    for j in range(i,5):
        t_stat, p_value = ttest_ind(locations[0][i], locations[0][j])
        if i != j:
            t_values.append(t_stat)
            print("T-test results for ",locations[1][i]," and ", locations[1][j],":")
            print("t-statistic:", t_stat)
            print("p-value:", p_value,"\n")
```

For all except two comparisons of prices in specified locations, the p-value is very low (approaching 0), therefore we reject the null hypothesis. This means there is strong evidence to conclude that there is a significant difference in median house values between these properties.

On the other hand, for properties (1H from Ocean vs Near Bay) and properties (Inland vs Island), the p value is greater than a certain threshhold (.05), which means we cannot reject the Null Hypothesis.

This hypothesis being: There is no difference in median house values between properties near the bay and those near the ocean.



```python
#Confidence level
confidence_level = 0.95
for i in range(5):
    #Sample mean and standard deviation
    mean_location = locations[0][i].mean()
    std_location = np.std(locations[0][i], ddof=1)  # ddof=1 for sample standard deviation
    
    #Number of observations in each group
    n_location = len(locations[0][i])
    
    #Degrees of freedom
    df_location = n_location - 1
    
    #Standard error of the mean for each group
    sem_location = std_location / np.sqrt(n_location)
    
    #Margin of error for each group
    margin_of_error_location = sem_location * stats.t.ppf((1 + confidence_level) / 2, df_location)
    
    #Confidence intervals for each group
    confidence_interval_location = (mean_location - margin_of_error_location, mean_location + margin_of_error_location)
    
    print("Confidence Interval for Median House Values", locations[1][i],":", confidence_interval_location,"\n")
```

The estimated population mean of median house values for properties in a specified area is likely to fall between the Confidence intervals above.
