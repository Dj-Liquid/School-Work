```python
import requests

response = requests.get('http://universities.hipolabs.com/search?country=India')
data = []
if response.status_code == 200:
    json_data = response.json()

    for item in json_data:
        name = item.get('name')
        state_province = item.get('state-province')
        if name and state_province:
            #print(f"Name: {name}, State/Province: {state_province}")
            data.append([name,state_province])
            

else:
    print('Error:', response.status_code)
#print(data)
df = pd.DataFrame(data,columns =['Name','State_Province'])
df.head()
```

```python
df.to_csv('University Data.csv')
```

```python
from bs4 import BeautifulSoup

response = requests.get('https://books.toscrape.com/catalogue/category/books/mystery_3/page-1.html')
soup = BeautifulSoup(response.text, 'html.parser')
books = soup.find_all('li', class_='col-xs-6 col-sm-4 col-md-3 col-lg-3')

book_list = []
for book in books:
    title = book.find('h3').find('a')['title']
    price = book.find('p', class_='price_color').text.strip()[1:]  # Removing the currency symbol
    review_rating = book.find('p', class_='star-rating')['class'][1]
    review_score = {
            'One': 1,
            'Two': 2,
            'Three': 3,
            'Four': 4,
            'Five': 5
        }[review_rating]
    book_list.append([title, price, review_score])

#print(len(book_list))
response = requests.get('https://books.toscrape.com/catalogue/category/books/mystery_3/page-2.html')
soup = BeautifulSoup(response.text, 'html.parser')
books = soup.find_all('li', class_='col-xs-6 col-sm-4 col-md-3 col-lg-3')

for book in books:
    title = book.find('h3').find('a')['title']
    price = book.find('p', class_='price_color').text.strip()[1:]  # Removing the currency symbol
    review_rating = book.find('p', class_='star-rating')['class'][1]
    review_score = {
            'One': 1,
            'Two': 2,
            'Three': 3,
            'Four': 4,
            'Five': 5
        }[review_rating]
    book_list.append([title, price, review_score])


#print(len(book_list))
#print(book_list)
df = pd.DataFrame(book_list,columns=['Title','Price','Review'])
df.tail()
```
