```python
import mysql.connector
import pandas as pd

connection = mysql.connector.connect(
  host = "database-comp3162.cqb8o27401uq.us-east-2.rds.amazonaws.com",
  user = "comp3162user",
  password = "Password876",
  port = 3306,
  database = "comp3162"
)
db_array = []
print("Connected to MYSQL database successfully")
print("\n")

cursor = connection.cursor()
cursor.execute(" SELECT COLUMN_NAME FROM INFORMATION_SCHEMA.COLUMNS WHERE TABLE_NAME = 'Patients'")
column_names = [row[0] for row in cursor.fetchall()]

cursor.execute("SELECT * FROM Patients")
rows = cursor.fetchall()
for row in rows:
    formatted_row = [col for col in row]
    print(formatted_row)
    db_array.append(formatted_row)
    

cursor.close()
connection.close()

df = pd.DataFrame(db_array, columns =column_names) 
```

```python
df.fillna(df.select_dtypes(include='number').mean(), inplace=True)
df.fillna(df.mode().iloc[0], inplace=True)
```

```python
null_values = df.isnull()

if null_values.any().any():
    print("There are still null values in the DataFrame.")
else:
    print("There are no null values in the DataFrame.")
```
