```python
import pandas as pd
import numpy as np
#620150009
```

```python
df = pd.read_csv("2022worldcup.csv")
df.head()
```

```python
print(df.describe())
```

```python
df[df['Age'] > 25]
```

```python
df.sort_values(by='Gls', ascending=False)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.bar(df['Squad'], df['Gls'])
plt.xlabel('Squad')
plt.ylabel('Total Goals')
plt.title('Total Goals Scored by Each Squad')
plt.xticks(rotation=90)
plt.show()

plt.figure(figsize=(8, 6))
plt.scatter(df['Age'], df['Min'])
plt.xlabel('Average Age')
plt.ylabel('Minutes Played')
plt.title('Relationship between Average Age and Minutes Played')
plt.show()

plt.figure(figsize=(8, 6))
plt.boxplot(df['Min'])
plt.xlabel('Minutes Played')
plt.title('Distribution of Minutes Played')
plt.show()
```

```python
print("Team(s) with most minutes played:")
print(df[df['Min'] == df['Min'].max()])
```

```python
print("Teams that scored more than 10 goals and played more than 500 minutes:")
print(df[(df['Gls'] > 10) & (df['Min'] > 500)])
```

```python
avg_goals = np.mean(df['Gls'] / df['Min'] * 90)
std_goals = np.std(df['Gls'] / df['Min'] * 90)

print("Mean goals scored per minute (Gls90):", avg_goals)
print("Standard deviation of goals scored per minute (Gls90):", std_goals)
```
